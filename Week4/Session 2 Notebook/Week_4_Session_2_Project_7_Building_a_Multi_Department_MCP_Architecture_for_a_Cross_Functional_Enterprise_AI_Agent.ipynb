### 💡 Key Concepts

#### Model Context Protocol (MCP)

**MCP** is a standardized protocol for AI agents to discover and use tools:

1. **Server**: Department exposes tools through MCP server (e.g., Finance on port 8010)
2. **Client**: Agent connects to server and discovers available tools at runtime
3. **Tool Call**: Agent invokes tools through standardized protocol
4. **Response**: Server executes tool and returns results to agent
5. **Synthesis**: Agent combines results from multiple servers into unified answer

Traditional enterprise AI architecture:
```
Agent → Custom API 1 → Database 1
      → Custom API 2 → Database 2
      → Custom API 3 → Database 3
(Requires building and maintaining custom integrations)
```

MCP-based architecture:
```
Agent → MCP Client → MCP Server 1 (Finance)
                   → MCP Server 2 (HR)
                   → MCP Server N (Other Depts)
(Standardized protocol, servers are plug-and-play)
```

#### RAG-Powered Tools

**RAG (Retrieval-Augmented Generation)** combines semantic search with LLM reasoning:
- Vector databases store embedded documents (department data, HR policies)
- Queries retrieve relevant context through similarity search
- LLMs use retrieved context to generate informed responses
- Persistent storage enables fast startup and reusability

This gives tools **access to large knowledge bases** without passing entire datasets to LLMs, reducing costs and improving accuracy.

#### Separation of Concerns

Each department maintains its own:
- **MCP Server**: Independent deployment on department infrastructure
- **Data Sources**: Department-controlled JSON files and databases
- **Vector Stores**: Persistent Chroma databases with department-specific embeddings
- **Business Logic**: Tools implement department-specific workflows

Agents remain **lightweight and generic**, only knowing how to connect to MCP servers — not requiring knowledge of any department's internal data or logic.

## 💰 Building the Finance MCP Server

### Creating a Finance Department MCP Server

This section creates a **Finance MCP Server** using FastMCP that provides intelligent budget analysis and spending approval recommendations.

**What is an MCP Server?**

MCP (Model Context Protocol) servers enable **any AI agent using an MCP client** to:
- Connect to this server at runtime
- Dynamically discover and pull available tools
- Use these finance capabilities without building custom tooling from scratch
- Access up-to-date department data and policies through the server's tools

This architecture separates **tool implementation** (MCP server) from **agent logic** (MCP client), allowing multiple agents across different teams to leverage the same finance capabilities.

**Real-World Deployment Scenario:**

In a production environment, imagine the **Finance Department** maintaining and deploying this MCP server on their own dedicated infrastructure:
- Finance team owns the server, data, and business logic
- Server runs on Finance department's secure servers (e.g., `finance.company.internal:8010`)
- Other departments' agents connect remotely to access finance tools
- Finance controls access, updates data, and maintains policies independently

> **Note:** In this Colab notebook, we simulate everything on the same server for demonstration purposes. In real enterprise scenarios, each department would deploy their MCP server on dedicated infrastructure, and agents would connect to multiple department servers across the network.

The server runs on **port 8010** with `mcp.run(transport="streamable-http")` to enable HTTP-based communication with agent clients.

### **Data Architecture**

The server loads department financial data from **`departments.json`** containing:
- **4 major departments**: Engineering, Sales, IT, Operations
- **Structured metadata**: budget, spent_ytd, headcount, fiscal year
- **Detailed financial narratives**: initiatives, spend breakdowns, risks, approval policies

**At startup, the server:**
1. Loads department data from JSON file
2. Creates LangChain Documents combining metadata with rich textual content
3. Initializes a **persistent Chroma vector database** (`./finance_db`) using OpenAI embeddings for semantic search
4. Configures a retriever with similarity threshold (`k=2, score_threshold=0.3`) to find relevant department information

### **Finance Advisory Tools**

The server exposes two intelligent finance tools that leverage RAG for context retrieval:

**1. `budget_analysis(query: str) -> str`**

Analyzes department budgets and spending patterns using natural language queries.

- **Purpose**: Answer questions about departmental finances, spending issues, and budget performance
- **Examples**:
  - "What's driving the cloud overspend in Engineering?"
  - "Show me Sales travel spending trends"
  - "How is IT performing against budget?"
- **How it works**: Retrieves relevant department documents via semantic search → LLM synthesizes analysis with specific metrics, insights, and recommendations
- **Returns**: Structured response with Analysis, Key Metrics, Insights, and Recommendations

**2. `spend_recommendation(spend_request: str) -> str`**

Provides approval recommendations for proposed spending requests based on department context and policies.

- **Purpose**: Evaluate spending requests and provide approval workflow guidance
- **Examples**:
  - "Engineering needs \$45,000 for AWS capacity for Kubernetes migration"
  - "Sales wants \$18,000 for conference travel to enterprise summit"
- **How it works**: Retrieves relevant department policies and budget status → LLM evaluates against approval thresholds and current spending
- **Returns**: Decision (APPROVE/REQUIRES_REVIEW/DENY), Required Approver, Rationale, and Next Steps

### **Server Initialization**

The server is initialized with `FastMCP(name="FinanceServer", instructions=...)` which:
- Documents available tools for agent discovery
- Specifies tool capabilities and expected inputs
- Enables the MCP client to intelligently route finance-related queries to appropriate tools

**Why FastMCP:**
- Simple decorator-based tool registration
- Built-in HTTP transport for easy deployment
- Automatic tool documentation for LLM agents
- Seamless integration with LangChain ecosystem

````python
%%writefile finance_mcp_server.py
from mcp.server.fastmcp import FastMCP
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_core.documents import Document
from langchain_chroma import Chroma
from dotenv import load_dotenv
import os
import json
from typing import List, Dict, Any

load_dotenv()

# ---------------------------------------------------------------------------
# Initialize LLM and Embeddings
# ---------------------------------------------------------------------------
llm = ChatOpenAI(
    model_name="gpt-4.1-mini",
    temperature=0.0,
    timeout=None,
)

embedder = OpenAIEmbeddings(
    model="text-embedding-3-small",
)

# ---------------------------------------------------------------------------
# Load Department Data from JSON
# ---------------------------------------------------------------------------
def load_departments_from_json(filepath: str = "departments.json") -> List[Dict[str, Any]]:
    """Load department data from JSON file."""
    try:
        with open(filepath, 'r') as f:
            department_data = json.load(f)
        print(f"✓ Loaded {len(department_data)} departments from {filepath}")
        return department_data
    except FileNotFoundError:
        print(f"✗ Error: {filepath} not found!")
        return []
    except json.JSONDecodeError as e:
        print(f"✗ Error parsing JSON: {e}")
        return []

# ---------------------------------------------------------------------------
# Create LangChain Documents from Department Data
# ---------------------------------------------------------------------------
def create_documents_from_departments(department_data: List[Dict[str, Any]]) -> List[Document]:
    """Convert department data into LangChain Documents for embedding.

    page_content: Combines dept_name, aliases, and financial_details for rich semantic search
    metadata: Structured fields for filtering and retrieval
    """
    documents = []

    for dept in department_data:
        # Build rich page_content for embedding
        aliases_text = ", ".join(dept.get("aliases", []))
        page_content = f"""Department: {dept['dept_name']}
Aliases: {aliases_text}

Financial Details:
{dept.get('financial_details', 'No financial details available')}
"""

        # Create metadata (exclude financial_details to avoid duplication)
        metadata = {
            "dept_name": dept["dept_name"],
            "fy": dept.get("fy", ""),
            "budget": dept.get("budget", 0),
            "spent_ytd": dept.get("spent_ytd", 0),
            "headcount": dept.get("headcount", 0),
            "aliases": dept.get("aliases", [])
        }

        doc = Document(page_content=page_content, metadata=metadata)
        documents.append(doc)

    print(f"✓ Created {len(documents)} documents for embedding")
    return documents

# ---------------------------------------------------------------------------
# Initialize Vector Store
# ---------------------------------------------------------------------------
def initialize_vector_store(documents: List[Document],
                            collection_name: str = "finance_departments",
                            persist_directory: str = "./finance_db") -> Chroma:
    """Create or load existing Chroma vector store from documents.

    If the collection already exists in the persist directory, it will be loaded.
    Otherwise, a new collection will be created and persisted.
    """
    try:
        # Try to load existing vector store
        vector_store = Chroma(
            collection_name=collection_name,
            embedding_function=embedder,
            persist_directory=persist_directory,
            collection_metadata={"hnsw:space": "cosine"}
        )

        # Check if collection has documents
        if vector_store._collection.count() > 0:
            print(f"✓ Loaded existing vector store with {vector_store._collection.count()} documents from {persist_directory}")
            return vector_store
        else:
            print(f"✓ Empty collection found, adding {len(documents)} documents...")
            vector_store.add_documents(documents)
            return vector_store

    except Exception as e:
        # If loading fails, create new vector store
        print(f"✓ Creating new vector store with {len(documents)} documents...")
        vector_store = Chroma.from_documents(
            documents=documents,
            embedding=embedder,
            collection_metadata={"hnsw:space": "cosine"},
            collection_name=collection_name,
            persist_directory=persist_directory
        )
        print(f"✓ Vector store initialized and persisted to {persist_directory}")
        return vector_store

# ---------------------------------------------------------------------------
# Load data and create vector store at startup
# ---------------------------------------------------------------------------
departments_data = load_departments_from_json("departments.json")
if not departments_data:
    raise RuntimeError("Failed to load department data. Server cannot start.")

finance_documents = create_documents_from_departments(departments_data)
vector_store = initialize_vector_store(
                    documents=finance_documents,
                    collection_name="finance_departments",
                    persist_directory="./finance_db"
)
db_retriever = vector_store.as_retriever(
            search_type="similarity_score_threshold",
            search_kwargs={"k": 2, "score_threshold": 0.3}
)

# ---------------------------------------------------------------------------
# Helper Function: Retrieve Context from Vector DB
# ---------------------------------------------------------------------------
def retrieve_finance_context(query: str) -> List[Document]:
    """Retrieve top-k relevant finance documents with error handling."""
    try:

        results = db_retriever.invoke(query)
        return results if results else []
    except Exception as e:
        print(f"✗ Error retrieving documents: {e}")
        return []

# ---------------------------------------------------------------------------
# MCP Server Setup
# ---------------------------------------------------------------------------
mcp = FastMCP(
    name="FinanceServer",
    instructions="""
Finance tools with RAG over detailed departmental budget documents.

AVAILABLE TOOLS:
1. budget_analysis(query): Analyze department budgets using natural language queries
2. spend_recommendation(amount, category, context): Get approval guidance for proposed spending

All responses are grounded in retrieved department financial data.
Departments available: Engineering, Sales, IT, Operations
""",
    port=8010
)

# ---------------------------------------------------------------------------
# Tool 1: Budget Analysis (Pure RAG)
# ---------------------------------------------------------------------------
@mcp.tool()
def budget_analysis(query: str) -> str:
    """Analyze department budgets and spending using natural language queries.

    Examples:
    - "What's the cloud spend issue in infrastructure?"
    - "Why is the data platform team under budget?"
    - "Show me Engineering headcount and spending"
    - "What are the risks in Sales travel spending?"
    - "How is Operations performing against budget?"

    Args:
        query: Natural language question about any department's budget, spending, or initiatives

    Returns:
        Detailed budget analysis with key metrics and actionable insights
    """
    # Retrieve top relevant documents from vector store
    docs = retrieve_finance_context(query)

    if not docs:
        return "I couldn't find any relevant department information for your query. Please try rephrasing or mention a specific department like Engineering, Sales, IT, or Operations."

    # Build context from all retrieved documents
    context = "RETRIEVED DEPARTMENT INFORMATION:\n\n"
    for i, doc in enumerate(docs, 1):
        dept_name = doc.metadata.get('dept_name', 'Unknown')
        context += f"--- DEPARTMENT {i}: {dept_name} ---\n"
        context += f"FY: {doc.metadata.get('fy', 'N/A')} | "
        context += f"Budget: ${doc.metadata.get('budget', 0):,} | "
        context += f"Spent YTD: ${doc.metadata.get('spent_ytd', 0):,} | "
        context += f"Headcount: {doc.metadata.get('headcount', 0)}\n\n"
        context += f"{doc.page_content}\n\n"

    # Generate analysis with LLM using all retrieved context
    prompt = f"""You are a Finance Business Partner providing budget analysis.

USER QUERY: {query}

{context}

INSTRUCTIONS:
1. Identify which department(s) from the retrieved information are most relevant to the query
2. Focus your analysis on the most relevant department, but mention others if they provide useful context
3. Extract specific numbers, initiatives, and risks from the retrieved information
4. Provide actionable insights based on the data

Provide a clear, actionable response in this format:

ANALYSIS:
[2-3 sentences directly answering the query, identifying the relevant department]

KEY METRICS:
- [Bullet points with specific numbers from the retrieved data]

INSIGHTS:
[1-2 important observations about trends, risks, or opportunities]

RECOMMENDATIONS:
[1-2 actionable next steps]

Keep the response under 200 words. Be specific with dollar amounts and percentages.
If multiple departments are relevant, clarify which one you're focusing on and why.
"""

    response = llm.invoke(prompt).content
    return response

# ---------------------------------------------------------------------------
# Tool 2: Spend Recommendation (Pure RAG)
# ---------------------------------------------------------------------------
@mcp.tool()
def spend_recommendation(spend_request: str) -> str:
    """Get approval recommendation for a proposed spend with proper workflow guidance.

    Provide details about the spend in natural language including:
    - Dollar amount
    - What it's for (category like cloud, travel, software, hardware, contractor)
    - Which team/department needs it
    - Why it's needed

    Examples:
    - "Engineering needs $45,000 for additional AWS capacity for Kubernetes migration"
    - "Sales rep wants to spend $8,000 on travel to an enterprise summit"
    - "IT department wants $150,000 for new collaboration software"
    - "Operations needs $25,000 for a temporary contractor to cover warehouse staffing"

    Args:
        spend_request: Natural language description of the spend request

    Returns:
        Approval decision with required approver, rationale, and policy reference
    """
    # Search for relevant department info and approval policies
    docs = retrieve_finance_context(spend_request)

    if not docs:
        return f"Cannot provide recommendation - no relevant department information found for this request."

    # Build comprehensive context from retrieved documents
    dept_context = "RETRIEVED DEPARTMENT INFORMATION:\n\n"
    for i, doc in enumerate(docs, 1):
        dept_name = doc.metadata.get('dept_name', 'Unknown')
        budget = doc.metadata.get('budget', 0)
        spent = doc.metadata.get('spent_ytd', 0)

        dept_context += f"--- DEPARTMENT {i}: {dept_name} ---\n"
        dept_context += f"Budget: ${budget:,} | Spent YTD: ${spent:,} | Remaining: ${budget - spent:,}\n\n"
        dept_context += f"{doc.page_content[:1200]}\n\n"

    # Generate recommendation
    prompt = f"""You are a Finance Operations specialist providing spend approval guidance.

SPEND REQUEST:
{spend_request}

{dept_context}

INSTRUCTIONS:
1. Extract the amount, category, and department from the spend request
2. Identify which department from the retrieved information is most relevant
3. Check that department's current budget status and spending patterns
4. Look for approval policies and thresholds mentioned in the retrieved information
5. Consider any risks, overspend situations, or relevant initiatives mentioned
6. Make a practical decision based on the amount, department context, and policies found

Provide your recommendation in this EXACT format:

DECISION: [APPROVE / REQUIRES_REVIEW / DENY]

RELEVANT DEPARTMENT: [which department from retrieved info]

REQUIRED APPROVER: [specific role/title based on policies found, or suggest appropriate level]

RATIONALE:
[2-3 sentences explaining your decision based on:
 - The department's budget status (over/under budget?)
 - if over or under budget what is the allocated budget, how much has been spent already and how much is remaining
 - Approval thresholds found in the retrieved policies
 - Whether this aligns with their initiatives or addresses known issues
 - Any relevant risks mentioned]

NEXT STEPS:
[1-2 specific actions, like "Submit PO to Director" or "Request pre-approval from VP"]

Be practical and reference specific information from the retrieved department data.
If the department is already over budget, be conservative. If this addresses a known issue or initiative, mention that.
"""

    response = llm.invoke(prompt).content
    return response

# ---------------------------------------------------------------------------
# Start the Server
# ---------------------------------------------------------------------------
if __name__ == "__main__":
    print("\n" + "="*70)
    print("🚀 Starting Finance MCP Server")
    print("="*70)
    print(f"📊 Departments loaded: {len(departments_data)}")
    print(f"📄 Documents embedded: {len(finance_documents)}")
    print(f"🔧 Tools available: budget_analysis, spend_recommendation")
    print(f"🌐 Server URL: http://127.0.0.1:8010/mcp")
    print("="*70 + "\n")

    mcp.run(transport="streamable-http")
````

## 👥 Building the HR MCP Server

### Creating an HR Department MCP Server

This section creates an **HR MCP Server** using FastMCP that provides intelligent PTO approval decisions and policy question answering.

**What is an MCP Server?**

MCP (Model Context Protocol) servers enable **any AI agent using an MCP client** to:
- Connect to this server at runtime
- Dynamically discover and pull available tools
- Use these HR capabilities without building custom tooling from scratch
- Access up-to-date employee data and HR policies through the server's tools

This architecture separates **tool implementation** (MCP server) from **agent logic** (MCP client), allowing multiple agents across different teams to leverage the same HR capabilities.

**Real-World Deployment Scenario:**

In a production environment, imagine the **HR Department** maintaining and deploying this MCP server on their own dedicated infrastructure:
- HR team owns the server, employee data, and policies
- Server runs on HR department's secure servers (e.g., `hr.company.internal:8011`)
- Other departments' agents connect remotely to access HR tools
- HR controls access, updates policies, and maintains employee records independently

> **Note:** In this Colab notebook, we simulate everything on the same server for demonstration purposes. In real enterprise scenarios, each department would deploy their MCP server on dedicated infrastructure, and agents would connect to multiple department servers across the network.

The server runs on **port 8011** with `mcp.run(transport="streamable-http")` to enable HTTP-based communication with agent clients.

### **Data Architecture**

The server uses a **hybrid approach** combining direct lookup and semantic search:

**1. Employee Data (`employees.json`) - Direct Dictionary Lookup**
- 5 US-based employees (EMP101-EMP105)
- Fields: employee_id, name, role, department, location, PTO balance, critical project status, tenure, manager
- Loaded into dictionary at startup for O(1) lookup by employee_id
- No vector database needed for employee records

**2. HR Policies (`hr_policies.json`) - Semantic Search**
- 5 comprehensive policies covering PTO, remote work, probation, conduct, and carryover rules
- Creates LangChain Documents with policy content
- Initializes **persistent Chroma vector database** (`./hr_db`) using OpenAI embeddings
- Retriever configured with similarity threshold (`k=3, score_threshold=0.3`)

**At startup, the server:**
1. Loads employee data into dictionary for fast lookup
2. Loads HR policies from JSON file
3. Creates LangChain Documents from policy data
4. Initializes persistent vector database for policy retrieval
5. Configures retriever for semantic search over policies

### **HR Advisory Tools**

The server exposes two intelligent HR tools combining direct lookup with semantic search:

**1. `pto_decision(employee_id: str, request_details: str) -> str`**

Analyzes PTO requests and provides approval decisions based on employee data and HR policies.

- **Purpose**: Evaluate PTO requests and provide approval workflow guidance with policy-based rationale
- **Hybrid Approach**:
  - Direct lookup of employee data (PTO balance, tenure, critical project status)
  - Semantic search over HR policies (approval rules, blackout periods, probation restrictions)
- **Examples**:
  - `employee_id="EMP101"`, `request_details='{"dates":"2025-09-02 to 2025-09-06","reason":"family event","backup":"Alice"}'`
  - `employee_id="EMP104"`, `request_details='{"dates":"2025-12-20 to 2025-12-24","reason":"vacation"}'`
- **How it works**: Looks up employee record → Retrieves relevant HR policies → LLM evaluates eligibility against policies → Generates decision
- **Returns**: Decision (APPROVE/APPROVE WITH CONDITIONS/DENY), Rationale, and Next Steps

**2. `hr_policy_question(question: str) -> str`**

Answers HR policy questions using company policies and guidelines.

- **Purpose**: Provide quick, accurate answers to employee policy questions
- **Examples**:
  - "Can I carry forward unused PTO to next year?"
  - "What's the remote work policy for Finance employees?"
  - "How much PTO do employees accrue per month?"
  - "Can probationary employees take PTO?"
- **How it works**: Retrieves relevant policy documents via semantic search → LLM synthesizes clear answer with citations
- **Returns**: Concise answer (2-5 sentences) with cited policy names

### **Server Initialization**

The server is initialized with `FastMCP(name="HRServer", instructions=...)` which:
- Documents available tools for agent discovery
- Specifies tool capabilities and expected inputs
- Enables the MCP client to intelligently route HR-related queries to appropriate tools

**Why FastMCP:**
- Simple decorator-based tool registration
- Built-in HTTP transport for easy deployment
- Automatic tool documentation for LLM agents
- Seamless integration with LangChain ecosystem

````python
%%writefile hr_mcp_server.py
from mcp.server.fastmcp import FastMCP
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_core.documents import Document
from langchain_chroma import Chroma
from dotenv import load_dotenv
import os
import json
from typing import List, Dict, Any

load_dotenv()

# ---------------------------------------------------------------------------
# Initialize LLM and Embeddings
# ---------------------------------------------------------------------------
llm = ChatOpenAI(
    model_name="gpt-4.1-mini",
    temperature=0.0,
    timeout=None,
)

embedder = OpenAIEmbeddings(
    model="text-embedding-3-small",
)

# ---------------------------------------------------------------------------
# Load Employee Data from JSON (Direct Lookup - No Vector DB)
# ---------------------------------------------------------------------------
def load_employees_from_json(filepath: str = "employees.json") -> Dict[str, Dict[str, Any]]:
    """Load employee data from JSON file and create dict keyed by employee_id."""
    try:
        with open(filepath, 'r') as f:
            employees_list = json.load(f)

        # Convert list to dict keyed by employee_id for O(1) lookup
        employees_dict = {emp["employee_id"]: emp for emp in employees_list}
        print(f"✓ Loaded {len(employees_dict)} employees from {filepath}")
        return employees_dict
    except FileNotFoundError:
        print(f"✗ Error: {filepath} not found!")
        return {}
    except json.JSONDecodeError as e:
        print(f"✗ Error parsing JSON: {e}")
        return {}

# ---------------------------------------------------------------------------
# Load HR Policy Data from JSON
# ---------------------------------------------------------------------------
def load_policies_from_json(filepath: str = "hr_policies.json") -> List[Dict[str, Any]]:
    """Load HR policy data from JSON file."""
    try:
        with open(filepath, 'r') as f:
            policies = json.load(f)
        print(f"✓ Loaded {len(policies)} HR policies from {filepath}")
        return policies
    except FileNotFoundError:
        print(f"✗ Error: {filepath} not found!")
        return []
    except json.JSONDecodeError as e:
        print(f"✗ Error parsing JSON: {e}")
        return []

# ---------------------------------------------------------------------------
# Create LangChain Documents from HR Policies
# ---------------------------------------------------------------------------
def create_documents_from_policies(policies: List[Dict[str, Any]]) -> List[Document]:
    """Convert HR policy data into LangChain Documents for embedding.

    page_content: Combines policy_name and details for rich semantic search
    metadata: Structured fields for filtering and retrieval
    """
    documents = []

    for policy in policies:
        # Build rich page_content for embedding
        page_content = f"""Policy: {policy['policy_name']}
Category: {policy.get('category', 'general')}

Details:
{policy.get('details', 'No details available')}
"""

        # Create metadata
        metadata = {
            "policy_name": policy["policy_name"],
            "category": policy.get("category", "general")
        }

        doc = Document(page_content=page_content, metadata=metadata)
        documents.append(doc)

    print(f"✓ Created {len(documents)} documents for embedding")
    return documents

# ---------------------------------------------------------------------------
# Initialize Vector Store
# ---------------------------------------------------------------------------
def initialize_vector_store(documents: List[Document],
                            collection_name: str = "hr_policies",
                            persist_directory: str = "./hr_db") -> Chroma:
    """Create or load existing Chroma vector store from documents.

    If the collection already exists in the persist directory, it will be loaded.
    Otherwise, a new collection will be created and persisted.
    """
    try:
        # Try to load existing vector store
        vector_store = Chroma(
            collection_name=collection_name,
            embedding_function=embedder,
            persist_directory=persist_directory,
            collection_metadata={"hnsw:space": "cosine"}
        )

        # Check if collection has documents
        if vector_store._collection.count() > 0:
            print(f"✓ Loaded existing vector store with {vector_store._collection.count()} documents from {persist_directory}")
            return vector_store
        else:
            print(f"✓ Empty collection found, adding {len(documents)} documents...")
            vector_store.add_documents(documents)
            return vector_store

    except Exception as e:
        # If loading fails, create new vector store
        print(f"✓ Creating new vector store with {len(documents)} documents...")
        vector_store = Chroma.from_documents(
            documents=documents,
            embedding=embedder,
            collection_metadata={"hnsw:space": "cosine"},
            collection_name=collection_name,
            persist_directory=persist_directory
        )
        print(f"✓ Vector store initialized and persisted to {persist_directory}")
        return vector_store

# ---------------------------------------------------------------------------
# Load data and create vector store at startup
# ---------------------------------------------------------------------------
EMPLOYEES_DICT = load_employees_from_json("employees.json")
if not EMPLOYEES_DICT:
    raise RuntimeError("Failed to load employee data. Server cannot start.")

policies_data = load_policies_from_json("hr_policies.json")
if not policies_data:
    raise RuntimeError("Failed to load HR policy data. Server cannot start.")

policy_documents = create_documents_from_policies(policies_data)
vector_store = initialize_vector_store(
    documents=policy_documents,
    collection_name="hr_policies",
    persist_directory="./hr_db"
)

db_retriever = vector_store.as_retriever(
    search_type="similarity_score_threshold",
    search_kwargs={"k": 3, "score_threshold": 0.3}
)

# ---------------------------------------------------------------------------
# Helper Function: Retrieve Policy Context from Vector DB
# ---------------------------------------------------------------------------
def retrieve_policy_context(query: str) -> List[Document]:
    """Retrieve top-k relevant HR policy documents with error handling."""
    try:
        results = db_retriever.invoke(query)
        return results if results else []
    except Exception as e:
        print(f"✗ Error retrieving documents: {e}")
        return []

# ---------------------------------------------------------------------------
# MCP Server Setup
# ---------------------------------------------------------------------------
mcp = FastMCP(
    name="HRServer",
    instructions="""
HR tools with employee lookup and RAG over HR policy documents.

AVAILABLE TOOLS:
1. pto_decision(employee_id, request_details): Analyze PTO requests and provide approval guidance
2. hr_policy_question(question): Answer HR policy questions using company policies

All responses are grounded in employee data and retrieved HR policy documents.
Employees available: EMP101-EMP105
Policies available: PTO/Leave, Remote Work, Probation, Conduct/Grievance, Carryover/Encashment
""",
    port=8011
)

# ---------------------------------------------------------------------------
# Tool 1: PTO Decision (Hybrid: Direct Employee Lookup + RAG for Policies)
# ---------------------------------------------------------------------------
@mcp.tool()
def pto_decision(employee_id: str, request_details: str) -> str:
    """Analyze PTO request and provide approval decision with policy-based rationale.

    Examples:
    - employee_id="EMP101", request_details='{"dates":"2025-09-02 to 2025-09-06","reason":"family event","backup":"Alice"}'
    - employee_id="EMP104", request_details='{"dates":"2025-12-20 to 2025-12-24","reason":"vacation","backup":"none","is_emergency":false}'

    Args:
        employee_id: Employee ID (e.g., EMP101, EMP102)
        request_details: JSON string or natural language describing the PTO request including dates, reason, backup coverage

    Returns:
        Decision (APPROVE / APPROVE WITH CONDITIONS / DENY) with rationale and next steps
    """
    # Step 1: Direct lookup of employee data
    employee = EMPLOYEES_DICT.get(employee_id)

    if not employee:
        return f"[HR] Error: Employee ID '{employee_id}' not found in system. Please verify the employee ID and try again."

    # Step 2: Retrieve relevant HR policies using RAG
    rag_query = f"PTO leave approval critical project probation {employee['role']} {employee['department']}"
    policy_docs = retrieve_policy_context(rag_query)

    if not policy_docs:
        return f"[HR] Warning: Could not retrieve relevant HR policies. Decision cannot be made without policy context."

    # Step 3: Build context from retrieved policies
    policy_context = "RETRIEVED HR POLICIES:\n\n"
    for i, doc in enumerate(policy_docs, 1):
        policy_name = doc.metadata.get('policy_name', 'Unknown Policy')
        policy_context += f"--- POLICY {i}: {policy_name} ---\n"
        policy_context += f"{doc.page_content}\n\n"

    # Step 4: Build employee context
    employee_context = f"""EMPLOYEE INFORMATION:
- Employee ID: {employee['employee_id']}
- Name: {employee['name']}
- Role: {employee['role']}
- Department: {employee['department']}
- Location: {employee['location']}
- PTO Balance: {employee['pto_balance_days']} days
- On Critical Project: {employee['on_critical_project']}
- Tenure: {employee['tenure_months']} months
- Manager: {employee['manager']}
- Exempt Status: {employee['exempt_status']}
"""

    # Step 5: Generate decision with LLM
    prompt = f"""You are an HR Business Partner making a PTO approval decision.

{employee_context}

PTO REQUEST DETAILS:
{request_details}

{policy_context}

INSTRUCTIONS:
1. Analyze the employee's eligibility for PTO based on:
   - PTO balance (do they have enough days?)
   - Tenure (are they in probationary period? <6 months = probation)
   - Critical project status (if on_critical_project=true, strict blackout rules apply)
   - Request duration (>3 days needs advance notice)

2. Check the retrieved policies for:
   - Approval thresholds and requirements
   - Critical project blackout restrictions
   - Probationary period limitations
   - Backup coverage requirements

3. Make a practical decision: APPROVE, APPROVE WITH CONDITIONS, or DENY

4. Provide clear rationale referencing specific policy requirements

Provide your response in this EXACT format:

DECISION: [APPROVE / APPROVE WITH CONDITIONS / DENY]

RATIONALE:
[2-4 sentences explaining your decision based on:
 - Employee's PTO balance and eligibility
 - Tenure/probation status if relevant
 - Critical project restrictions if applicable
 - Policy requirements from retrieved documents
 - Any conditions that must be met]

NEXT STEPS:
1. [Specific action for employee]
2. [Specific action for manager]
3. [Optional: Additional requirement or follow-up]

Be professional, clear, and reference specific policy requirements. If denying, suggest alternatives if possible.
"""

    response = llm.invoke(prompt).content
    return response

# ---------------------------------------------------------------------------
# Tool 2: HR Policy Question (Pure RAG)
# ---------------------------------------------------------------------------
@mcp.tool()
def hr_policy_question(question: str) -> str:
    """Answer HR policy questions using company HR policies and guidelines.

    Examples:
    - "Can I carry forward unused PTO to next year?"
    - "What's the remote work policy for Finance employees?"
    - "How much PTO do employees accrue per month?"
    - "What happens if I'm absent without notice for 3 days?"
    - "Can probationary employees take PTO?"

    Args:
        question: Natural language question about HR policies, benefits, or workplace rules

    Returns:
        Clear answer with relevant policy citations
    """
    # Retrieve relevant policy documents
    policy_docs = retrieve_policy_context(question)

    if not policy_docs:
        return "I couldn't find relevant HR policy information to answer your question. Please try rephrasing or contact HR directly for assistance."

    # Build context from retrieved policies
    context = "RETRIEVED HR POLICY INFORMATION:\n\n"
    policy_names = []

    for i, doc in enumerate(policy_docs, 1):
        policy_name = doc.metadata.get('policy_name', 'Unknown Policy')
        policy_names.append(policy_name)
        context += f"--- POLICY {i}: {policy_name} ---\n"
        context += f"{doc.page_content}\n\n"

    # Generate answer with LLM
    prompt = f"""You are an HR policy assistant helping employees understand company policies.

EMPLOYEE QUESTION:
{question}

{context}

INSTRUCTIONS:
1. Provide a clear, concise answer to the employee's question (2-5 sentences)
2. Reference specific policy requirements, thresholds, or rules from the retrieved information
3. If there are important conditions or exceptions, mention them
4. If the question requires manager approval or HR consultation, state that clearly
5. Use professional but friendly language

At the end, add:
Cited Policies: [comma-separated list of policy names used]

Keep the response under 200 words. Be accurate and helpful.
"""

    response = llm.invoke(prompt).content

    # Ensure citation line exists
    if "Cited Policies:" not in response and "cited policies:" not in response.lower():
        cited = ", ".join(policy_names) if policy_names else "N/A"
        response = f"{response.rstrip()}\n\nCited Policies: {cited}"

    return response

# ---------------------------------------------------------------------------
# Start the Server
# ---------------------------------------------------------------------------
if __name__ == "__main__":
    print("\n" + "="*70)
    print("🚀 Starting HR MCP Server")
    print("="*70)
    print(f"👥 Employees loaded: {len(EMPLOYEES_DICT)}")
    print(f"📋 Policies loaded: {len(policies_data)}")
    print(f"📄 Documents embedded: {len(policy_documents)}")
    print(f"🔧 Tools available: pto_decision, hr_policy_question")
    print(f"🌐 Server URL: http://127.0.0.1:8011/mcp")
    print("="*70 + "\n")

    mcp.run(transport="streamable-http")
````

## 🤖 Building the Cross-Functional AI Agent

### Creating the MCP Client Agent

This section creates a **Cross-Functional AI Agent** that connects to multiple department MCP servers and orchestrates tool usage across Finance and HR.

**What is an MCP Client Agent?**

An MCP client agent acts as an intelligent orchestrator that:
- **Discovers tools** from multiple MCP servers at runtime using MCP
- **Routes queries** to the appropriate department based on user intent
- **Executes tools** by calling the relevant MCP server endpoints
- **Synthesizes responses** from multiple department tools into coherent answers

This creates a unified interface where users can ask cross-departmental questions without knowing which server to contact.

**Agent Architecture:**

The agent uses **LangGraph's React Agent** pattern:
- **Model**: GPT-4.1-mini with temperature=0 for deterministic reasoning
- **Tools**: Dynamically loaded from Finance (port 8010) and HR (port 8011) MCP servers
- **System Prompt**: Guides the agent on how to use tools and format responses
- **Streaming**: Real-time response streaming for better user experience

**Multi-Server Connection:**

The agent connects to **two department MCP servers** simultaneously using a **multi-server MCP Client**:

| Server | Endpoint | Tools Available |
|--------|----------|-----------------|
| Finance | `http://127.0.0.1:8010/mcp` | `budget_analysis`, `spend_recommendation` |
| HR | `http://127.0.0.1:8011/mcp` | `pto_decision`, `hr_policy_question` |

**How the Agent Works:**

1. **Tool Discovery**: Agent connects to both MCP servers and retrieves available tools
2. **Query Analysis**: User asks a question (e.g., "What's the cloud spend in Engineering?")
3. **Tool Selection**: Agent analyzes the query and selects appropriate tool (`budget_analysis`)
4. **Tool Execution**: Agent calls the Finance MCP server with the query
5. **Response Synthesis**: Agent formats the tool response into a user-friendly answer
6. **Streaming Output**: Response is streamed to the user in real-time with rich formatting

**Test Queries:**

The agent is configured to test all four tools with realistic scenarios:

1. **Finance Budget Analysis**: "What's driving the cloud overspend in Engineering?"
   - Tests: Semantic search over department data, budget analysis capabilities

2. **Finance Spend Approval**: "Should we approve $45,000 for AWS capacity for Kubernetes migration?"
   - Tests: Policy retrieval, approval workflow guidance, budget context

3. **HR PTO Decision**: "Decide on PTO for EMP101 from Sept 2-6 for family event with backup Alice"
   - Tests: Employee lookup, policy retrieval, approval decision logic

4. **HR Policy Question**: "Can I carry forward unused PTO to next year?"
   - Tests: Semantic search over HR policies, policy citation

**Visual Output:**

The agent uses **Rich library** for professional terminal output:
- 🔧 Tool discovery display with color-coded tool names
- 📋 Query panels with borders and titles
- 🤖 Processing indicators during tool execution
- ✅ Final response panels with Markdown formatting
- Progress tracking (Query 1/4, 2/4, etc.)

**Why This Architecture:**

- **Separation of Concerns**: Departments maintain their own servers and data
- **Scalability**: Easy to add new department servers without modifying the agent
- **Security**: Each department controls access to their own tools
- **Flexibility**: Agents can be deployed anywhere and connect to multiple servers
- **Maintainability**: Department logic stays with departments, not in the agent

> **Real-World Usage:** In production, this agent could be deployed as a Slack bot, web interface, or API endpoint, allowing employees across the organization to ask cross-departmental questions through a single interface.

````python
%%writefile client_agent.py
from langchain_mcp_adapters.client import MultiServerMCPClient
from langchain.agents import create_agent
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage
import asyncio
from dotenv import load_dotenv
import os
from agent_utils import format_message
from rich.console import Console
from rich.markdown import Markdown
from rich.panel import Panel
from rich.text import Text

console = Console()
load_dotenv()

# Initialize the LLM with deterministic behavior
llm = ChatOpenAI(
    model_name="gpt-4.1-mini",
    temperature=0.0,
    timeout=None,
)

SYSTEM_PROMPT = """You are an AI assistant with access to Finance and HR management tools via MCP servers.

RESPONSE GUIDELINES:
1. Call the appropriate tools based on the user's query
2. Provide detailed, comprehensive answers based on the information returned by tools
3. Preserve formatting like bullet points and structured data from tool responses
4. Reference specific numbers, policies, and data points from tool results if available
5. If a tool returns an error, clearly communicate it to the user
6. For approval decisions, include the decision, rationale, and next steps
7. For policy questions, cite the relevant policies used

Be professional, accurate, and helpful. Always ground your responses in the data returned by the tools.
"""

async def main():
    # Initialize MCP client with both servers
    client = MultiServerMCPClient({
        "finance": {
            "url": "http://127.0.0.1:8010/mcp",
            "transport": "streamable_http",
        },
        "hr": {
            "url": "http://127.0.0.1:8011/mcp",
            "transport": "streamable_http",
        },
    })

    # Get available tools
    tools = await client.get_tools()

    # Display discovered tools
    console.print("\n[bold green]🔧 Discovered MCP Tools:[/bold green]")
    for tool in tools:
        console.print(f"  • [green]{tool.name}[/green]")
    console.print()

    # Create agent
    agent = create_agent(model=llm, tools=tools, system_prompt=SYSTEM_PROMPT)

    # Test queries
    samples = [
        # Finance: Budget analysis
        "Finance: What's driving the cloud overspend in Engineering?",

        # Finance: Spend recommendation
        "Finance: We need to spend $45,000 on additional AWS capacity for the Kubernetes migration in the infrastructure team. Should we approve this?",

        # HR: PTO decision for critical project employee
        'HR: Decide on PTO for EMP101, with the request of {"dates":"2025-09-02 to 2025-09-06","reason":"family event","backup":"Alice"}',

        # HR: Policy question
        "HR policy: Can I carry forward unused PTO to next year?"
    ]

    # Process each query
    for idx, query in enumerate(samples, 1):
        # Display query header
        console.print("\n" + "="*80)
        console.print(Panel(
            Text(query, style="bold green"),
            title=f"[bold red]Query {idx}/{len(samples)}[/bold red]",
            border_style="red"
        ))

        # Stream agent response
        console.print("\n[bold red]🤖 Agent Processing:[/bold red]")

        final_event = None
        async for event in agent.astream(
            input={"messages": [HumanMessage(content=query)]},
            config={"recursion_limit": 25},
            stream_mode="values"
        ):
            if event:
                format_message(event["messages"][-1])
                final_event = event

        # Display final formatted response
        if final_event:
            console.print("\n[bold green]✅ Final Response:[/bold green]")
            console.print(Panel(
                Markdown(final_event["messages"][-1].content),
                border_style="green",
                padding=(1, 2)
            ))

        console.print()

    console.print("="*80)
    console.print("[bold cyan]✨ All queries completed![/bold cyan]\n")


if __name__ == "__main__":
    asyncio.run(main())
````
