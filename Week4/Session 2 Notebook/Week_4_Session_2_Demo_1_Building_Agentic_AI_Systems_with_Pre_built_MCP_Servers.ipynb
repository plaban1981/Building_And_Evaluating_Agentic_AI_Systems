### 💡 Key Concepts

#### Model Context Protocol (MCP)

**MCP** is a universal standard for AI agents to discover and use tools:

1. **Server**: Exposes capabilities as tools through MCP endpoint (e.g., `http://localhost:8011/mcp`)
2. **Client**: Agent connects to server and discovers available tools
3. **Tool Call**: Agent invokes tools using standardized protocol
4. **Response**: Server executes tool and returns results
5. **Synthesis**: Agent uses results to complete user task

Traditional agent tool development:
```
Agent → Custom Python Function → External Library → Result
(Developer writes and maintains wrapper code for each tool)
```

MCP-based agent workflow:
```
Agent → MCP Client → MCP Server → Tool Implementation → Result
(Tool is pre-built, maintained by experts, used via standard protocol)
```

#### Pre-built vs Custom MCP Servers

**Pre-built MCP Servers** (this demo):
- Built and maintained by tool creators (Microsoft, LangChain, etc.)
- Ready to deploy with simple commands
- Production-tested and optimized
- Examples: Markitdown, LangChain Docs, Brave Search, GitHub, Slack

**Custom MCP Servers** (like Finance/HR in the next project):
- Built by your organization or different teams for internal tools and data
- Exposes company-specific capabilities
- Implements custom business logic
- Examples: Internal databases, proprietary APIs, company knowledge bases

#### Why Pre-built Servers Matter

Organizations and open-source projects are building MCP servers for common needs:
- **Document Processing**: Markitdown (Microsoft), Unstructured.io
- **Search**: Brave Search, Google Search, web scraping
- **Productivity**: Slack, Gmail, Google Calendar, Notion
- **Development**: GitHub, GitLab, Linear
- **Data**: PostgreSQL, MongoDB, Redis

Instead of building these integrations yourself, **use pre-built MCP servers** and focus on your core business logic.

## 🚀 Deploying the Pre-built Markitdown MCP Server

### What is the Markitdown MCP Server?

The **Markitdown MCP Server** is a pre-built Model Context Protocol (MCP) server that wraps Microsoft's **Markitdown** library — a powerful document conversion tool that transforms various document formats into clean, structured Markdown.

**Microsoft Markitdown** supports conversion from:
- **Documents**: PDF, DOCX, PPTX, XLSX
- **Web content**: HTML, URLs (web pages)
- **Media**: Images (with OCR), audio transcripts
- **Code**: Various programming language files
- **Other formats**: CSV, JSON, XML

The MCP server exposes this functionality as a standardized tool (`convert_to_markdown`) that AI agents can discover and use at runtime.

### Why Use a Pre-built MCP Server?

**Traditional approach (without MCP):**
- Write custom code to convert documents
- Handle multiple file formats separately
- Integrate conversion logic into your agent codebase
- Maintain and update conversion libraries yourself

**Pre-built MCP Server approach:**
- Microsoft has already built and maintains the conversion logic
- The MCP server exposes it as a standard tool
- Your agent discovers and uses it automatically
- No custom implementation needed

> **Key Advantage:** If Microsoft hosted this MCP server on their own infrastructure (e.g., `https://markitdown.microsoft.com/mcp`), we could simply point our agent to that URL. In this demo, we deploy the pre-built server **locally** to simulate the same experience without depending on external services.

### Deployment Configuration

We launch the Markitdown MCP server as a **background process** on our local machine:

**Server configuration:**
- **Command**: `markitdown-mcp` (pre-built server executable)
- **Transport**: `--http` (HTTP-based communication)
- **Host**: `127.0.0.1` (localhost)
- **Port**: `8011`
- **Logs**: Output written to `server_output.log`

**Command flags explained:**
- `nohup` - Keeps the server running even if terminal session closes
- `--http` - Enables HTTP transport (vs SSE or stdio)
- `--host 127.0.0.1` - Binds to localhost
- `--port 8011` - Listens on port 8011
- `> server_output.log 2>&1` - Redirects all output to log file
- `&` - Runs process in background

Once running, the server exposes the **`convert_to_markdown`** tool at `http://127.0.0.1:8011/mcp`, ready for agent discovery and use.

> **Real-World Scenario:** In production, organizations can deploy pre-built MCP servers like this on their infrastructure and provide URLs to teams. Agents across the organization can then use these shared services without reimplementing functionality.

### Creating an Agent that Connects to Multiple Pre-built MCP Servers

This section creates a **Multi-Server MCP Client Agent** that demonstrates the true power of the Model Context Protocol — connecting to multiple tool servers (both local and remote) and using their capabilities seamlessly.

**What Makes This Agent Special?**

- **Discovers tools dynamically** from multiple MCP servers at runtime
- **Connects to heterogeneous servers** (local Markitdown + remote LangChain Docs)
- **Routes tasks intelligently** to the appropriate tool based on the query
- **Requires zero custom tool implementation** — everything comes from MCP servers

This is the **standard MCP pattern**: the agent is a generic orchestrator that discovers and uses tools from any MCP-compliant server.

### Agent Tasks

The agent performs two distinct tasks that showcase both MCP servers:

- **Agent workflow**:

  3. Calls LangChain Docs server with search query

### Key Technical Patterns

**3. File Output**

A single agent can connect to all of them, giving users access to cross-functional capabilities through one unified interface. We will cover this in the next project.

````python
%%writefile client_agent.py

import asyncio
import os
from dotenv import load_dotenv

from langchain.agents import create_agent
from langchain_openai import ChatOpenAI
from langchain_mcp_adapters.client import MultiServerMCPClient
from langchain_core.messages import HumanMessage
from agent_utils import format_message

# Load environment variables (e.g., OPENAI_API_KEY)
load_dotenv()

# Use the correct model name
llm = ChatOpenAI(model="gpt-4.1-mini", temperature=0)

async def main():
    # Initialize MCP Client
    client = MultiServerMCPClient({
        "markitdown": {
            # Markitdown MCP Server:
            # Converts web pages and documents into structured Markdown
            "url": "http://localhost:8011/mcp",
            "transport": "streamable_http"
        },
        "langchain": {
            # LangChain documentation MCP server:
            # Enables semantic search over LangChain and LangGraph docs
            "url": "http://docs.langchain.com/mcp",
            "transport": "streamable_http"
        }
    })

    # Discover tools exposed by all connected MCP servers
    # This is where the agent learns what actions it can take
    try:
        tools = await client.get_tools()
        print(f"Discovered tools: {[tool.name for tool in tools]}")
    except Exception as e:
        print(f"Error connecting to MCP servers: {e}")
        return

    AGENT_PROMPT_TXT = """Act as a helpful assistant.
    Use the tools available to generate responses to queries.

    Specific Tool Context:
    - SearchDocsByLangChain: For answering user queries on langchain and langgraph.
    - convert_to_markdown: To extract info from websites to markdown reports.
    """

    agent = create_agent(
        model=llm,
        tools=tools,
        system_prompt=AGENT_PROMPT_TXT
    )


    # ---------------------------------------------------------
    # TASK 1: Website → Structured Markdown Extraction
    # ---------------------------------------------------------

    print('\n--- Task 1: Document Extraction ---')
    source = "https://www.ibm.com/think/topics/agentic-ai"
    prompt_1 = f"Create a report on the following article by focusing on key sections and content. Source: {source}"
    async for event in agent.astream(input={"messages": [HumanMessage(content=prompt_1)]},
                                         config={"recursion_limit": 25}, stream_mode="values"):
            if event:
                format_message(event["messages"][-1])


    # Extract the last message content (the final answer)
    content_1 = event["messages"][-1].content
    with open("document1.md", 'w') as f:
        f.write(content_1)
    print('Website Extraction completed -> document1.md\n')

    # ---------------------------------------------------------
    # TASK 2: LangChain Agent Tutorial Generation
    # ---------------------------------------------------------

    print('\n--- Task 2: LangChain Tutorial ---')
    prompt_2 = "How to create a react agent with langchain in python"
    async for event in agent.astream(input={"messages": [HumanMessage(content=prompt_2)]},
                                        config={"recursion_limit": 25}, stream_mode="values"):
        if event:
            format_message(event["messages"][-1])

    content_2 = event["messages"][-1].content
    with open("document2.md", 'w') as f:
        f.write(content_2)
    print('LangChain tutorial completed -> document2.md\n')

if __name__ == "__main__":
    asyncio.run(main())
````
