**Purpose:** Store and update user-specific memories proactively.

### Tool Binding to LLM

We bind all tools to the LLM, enabling **tool calling** capability:


## Create Tools for AI Agent

````python
from langmem import create_manage_memory_tool, create_search_memory_tool
from langchain.tools import tool
from langchain_community.utilities.tavily_search import TavilySearchAPIWrapper
from langchain_community.retrievers import ArxivRetriever
from langchain_openai import ChatOpenAI

# Tool for web search on general health topics
# Tavily Web Search
tavily_search = TavilySearchAPIWrapper()

@tool
def search_web(query: str) -> list:
    """
    Search the web for general or up-to-date information on healthcare topics.

    Inputs:
    - query (str): The search query string. Should describe the healthcare topic or information you want to find.

    Outputs:
    - list: A list of up to 3 formatted strings, each containing:
        - Title of the search result
        - Content extracted from the page
        - Source URL
    """
    results = tavily_search.raw_results(query=query, max_results=3, search_depth='advanced',
                                        include_answer=False, include_raw_content=True)
    docs = results['results']
    docs = [doc for doc in docs if doc.get("raw_content") is not None]
    docs = ['## Title\n'+doc['title']+'\n\n'+'## Content\n'+doc['raw_content']+'\n\n'+'##Source\n'+doc['url'] for doc in docs]
    return docs


namespace = ("agent_memories", "{user_id}")
MEMORY_MGMT_SYS_PROMPT = """Proactively call this tool when you:

1. Identify a new **USER preference**, especially related to:
   - Learning goals and skill development areas
   - Topics of interest and subject matter focus
   - Preferred teaching or mentoring style
   - Communication tone, level of detail, and pacing
   - Formats, examples, or resources the USER finds helpful
2. Receive an explicit USER request to **remember** something or adjust your behavior.
3. Are in an ongoing mentoring conversation and want to **record important context** that will help personalize future guidance.
4. ONLY UPDATE when an older memory needs replacing else keep ADD NEW MEMORIES based on new user informaiton as per point 1.
"""


create_mem_tool = create_manage_memory_tool(namespace=namespace,
                                            instructions=MEMORY_MGMT_SYS_PROMPT)
search_mem_tool = create_search_memory_tool(namespace)
````

````python
tools = [search_web, create_mem_tool, search_mem_tool]

llm = ChatOpenAI(model="gpt-4.1-mini", temperature=0)
llm_with_tools = llm.bind_tools(tools=tools)
````

### Create the Agent Graph with Long and Short Term memory

### Agent State Schema

We define a simple state schema that holds the conversation messages.

**State Design:**
```python
class State(TypedDict):
    messages: Annotated[list, add_messages]
```

**Key aspects:**
- **`messages`** вҖ” List of all conversation messages (Human, AI, Tool)
- **`Annotated[list, add_messages]`** вҖ” Tells LangGraph to merge new messages instead of replacing
- **Simple by design** вҖ” Memory lives in the checkpointer and store, not state

**Why this works:**
- вң… Clean separation: state = conversation, checkpointer = short-term memory, store = long-term memory
- вң… Easy to reason about message flow
- вң… Standard LangGraph pattern

````python
from typing import Annotated
from typing_extensions import TypedDict
from langgraph.graph.message import add_messages

class State(TypedDict):
    messages: Annotated[list, add_messages]
````

```
           вҶ“
      Tool Call?
       вҶҷ     вҶҳ
     Yes     No
      вҶ“       вҶ“
  [Tools]   END
      вҶ“
  [Agent] вҶҗ feedback loop
      вҶ“
     END
```

**Node Functions:**

**1. Agent Node (`tool_calling_llm`)**
```python
def tool_calling_llm(state: State) -> State:
    # Prepend system instructions to conversation
    state_with_instructions = [SYS_PROMPT] + state["messages"]
    
    # Invoke LLM (can return text or tool calls)
    response = llm_with_tools.invoke(state_with_instructions)
    
    # Add response to state
    return {"messages": [response]}
```

**2. Tools Node** (Prebuilt `ToolNode`)
- Automatically executes tool calls from agent
- Handles search_web, manage_memory, and search_memory
- Returns results as ToolMessage

**Conditional Routing:**
- Uses custom `agent_conditional_routing` routing function to check if LLM made tool calls
- Routes to `tools` if tool calls present
- Routes to `END` if text response

**Feedback Loop:**
- `tools вҶ’ agent` edge creates the reasoning loop
- Agent can make multiple tool calls in sequence
- Continues until agent produces final answer

**Memory Systems:**

**1. Short-Term Memory for each user-session (InMemorySaver)**
```python
short_term_memory = InMemorySaver()
```
- Stores conversation history per thread (user-session)
- Isolated by `thread_id`
- Enables multi-turn conversations per user-session

**2. Long-Term Memory for each user (InMemoryStore)**
```python
long_term_memory_store = InMemoryStore(
    index={
        "dims": 1536,  # OpenAI embedding dimensions
        "embed": "openai:text-embedding-3-small"  # Embedding model
    }
)
```
- Stores user-specific memories from multiple user-sessions for the same user
- Vector search with embeddings
- Isolated by user namespace

**Compilation:**
```python
agent = builder.compile(
    checkpointer=short_term_memory,  # Session context
    store=long_term_memory_store     # User memories
)
```

**Why this architecture works:**
- вң… Clean separation of concerns
- вң… Scalable to many users and sessions
- вң… Proper memory isolation
- вң… Production-ready patterns

````python
from langchain_core.messages import SystemMessage, HumanMessage
from langgraph.graph import StateGraph, START, END
from langgraph.prebuilt import ToolNode
from IPython.display import display, Image, Markdown
from langgraph.checkpoint.memory import InMemorySaver
from langgraph.store.memory import InMemoryStore
from typing import Literal

# Create the node function that handles reasoning and planning using the LLM
def tool_calling_llm(state: State) -> State:
    # Extract the current conversation history from the state
    current_state = state["messages"]

    # Prepend the system instructions to the current message history
    state_with_instructions = [SYS_PROMPT] + current_state

    # Call the LLM to generate a new message (either a response or a tool call request)
    response = [llm_with_tools.invoke(state_with_instructions)]

    # Return the updated state containing the new message
    return {"messages": response}

# Build the graph
builder = StateGraph(State)

# Add nodes
builder.add_node("agent", tool_calling_llm)
builder.add_node("tools", ToolNode(tools=tools))

# Add edges
builder.add_edge(START, "agent")

# Conditional routing function from agent to either tools or END
def agent_conditional_routing(state: State) -> Literal["goto_tools_route", "goto_end_route"]:
    """Decide whether to call tools or end."""
    last_state_message = state["messages"][-1] # access the most recent messages update by agent node

    if hasattr(last_state_message, 'tool_calls') and last_state_message.tool_calls and len(last_state_message.tool_calls) > 0:
        return "goto_tools_route"
    return "goto_end_route"

# add conditional edges
builder.add_conditional_edges(
    "agent",
    agent_conditional_routing, # conditional routing function
    {
        "goto_tools_route": "tools", # If the latest message (result) from LLM is a tool call request -> tools_condition routes to tools
        "goto_end_route": END # If the latest message (result) from LLM is a not a tool call -> tools_condition routes to END
    }
)
builder.add_edge("tools", "agent") # this is the key feedback loop in the agentic system

# Compile Agent Graph
short_term_memory = InMemorySaver()
long_term_memory_store = InMemoryStore(
    index={
        "dims": 1536,
        "embed": "openai:text-embedding-3-small"
    }
)
agent = builder.compile(checkpointer=short_term_memory,
                        store=long_term_memory_store)
````

### Execution Wrapper

We use a helper function to stream agent execution and display results:

### Run and Test the Agent

````python
from agent_utils import format_message

def call_mentor_agent(agent, query, user_id, session_id, verbose=False):

    for event in agent.stream(
        input={"messages": [{"role": "user", "content": query}]},
        config={"configurable": {"user_id": user_id, "thread_id": session_id}},
        stream_mode='values' #returns full agent state with all messages including updates
    ):
        if verbose:
            format_message(event["messages"][-1])

    print('\n\nFinal Response:\n')
    display(Markdown(event["messages"][-1].content.replace(r'$', r'\$')))
    return event["messages"]
````
